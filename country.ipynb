```python
import numpy as np
import pandas as pd
```

```python
population = pd.read_csv("/content/country_population.csv")
fertility = pd.read_csv("/content/fertility_rate.csv")
life = pd.read_csv("/content/life_expectancy.csv")
country = pd.read_csv("/content/Metadata_Country.csv")
```

```python
country.head()
```

```python
country = country[['Country Code','Region']]
country.head()
country.shape
```

```python
years = [str(i) for i in range(1960,2017)]
years
```

```python
population.drop(['Indicator Name','Indicator Code'],axis = 1, inplace = True)
```

```python
population.dropna(axis=0, inplace=True)
```

```python
population.shape
```

```python
df = pd.melt(population,
             id_vars ='Country Code',
             value_vars = years,
             var_name ='Year',
             value_name = 'Population'
             )

df.head()
```

```python
df_merged = pd.merge(country,df,how='left',on='Country Code')
df_merged.head()
```

```python
df_merged.tail()
```

```python
fertility.head()
```

```python
fertility.drop(['Indicator Name','Indicator Code'],axis = 1, inplace = True)
```

```python
fertility.head()
```

```python
fertility.dropna(axis=0, inplace=True)
```

```python
fertility.shape
```

```python
df1 = pd.melt(fertility,
             id_vars ='Country Code',
             value_vars = years,
             var_name ='Year',
             value_name = 'Fertility_Rate'
             )

df1.head()
```

```python
df2 = pd.merge(df1, df_merged, how='left', on=['Country Code', 'Year'])
df2.head()
```

```python
life.head()
```

```python
life.drop(['Indicator Name','Indicator Code'],axis = 1, inplace = True)
```

```python
life.dropna(axis=0, inplace=True)
```

```python
life.head()
```

```python
df3 = pd.melt(life,
             id_vars ='Country Code',
             value_vars = years,
             var_name ='Year',
             value_name = 'Life_Expectancy'
             )

df3.head()
```

```python
data = pd.merge(df2, df3, how='left', on=['Country Code', 'Year'])
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(axis=0, inplace=True)
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.head()
```

```python
data.tail()
```
